```python
dataset = pd.read_csv("/content/netflix_titles.csv")
```

```python
print("-------------top 10 rows-------------")
print(dataset.head(10))
print("-------------list of columns-------------")
print(dataset.info())
print("-------------empty values-------------")
print(dataset.isnull().sum())
print(dataset.describe())
```

```python
clean_dataset = dataset.dropna()
print(clean_dataset.isnull().sum())
print(clean_dataset.describe())
print(clean_dataset.duplicated().sum())
# clean_dataset.drop_duplicates(inplace=True)
# print(clean_dataset.describe())
```

```python
print(clean_dataset.groupby(by='country').count())
```

```python
print(clean_dataset[clean_dataset['country']=='India'].count())
print(clean_dataset[clean_dataset['release_year']==2021].count())
print(clean_dataset['release_year'].min())
```

```python
clean_dataset['rating'].value_counts().plot(kind='bar')

plt.xlabel('rating')
plt.ylabel('count')
plt.title('movies per rating')
plt.show()
```

```python
clean_dataset['release_year'].plot(kind='hist', bins=range(1940,2030,10))

plt.xlabel('decades')
plt.ylabel('number of titles')
plt.title('movies per decade')
plt.show()
```

```python
movies_per_year = clean_dataset['release_year'].value_counts().sort_index()

movies_per_year.plot(kind='line')

plt.xlabel('year')
plt.ylabel('count')
plt.title('movies per year')
plt.show()
```
